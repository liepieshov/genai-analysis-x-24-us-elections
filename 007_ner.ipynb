```python
import tweetnlp
import pandas as pd
from tqdm.notebook import tqdm
```

```python
df = pd.read_csv("data_small.csv")
df.dt = pd.to_datetime(df.dt)
df.user_created = pd.to_datetime(df.user_created)
df.head()
```

```python
model = tweetnlp.load_model('ner')
```

```python
# reversed
result = []
for x in tqdm(df['text'].values[::-1]):
    result.append(model.ner(x, return_probability=True))
```
