```python
result_df = pd.read_csv("data_topic.csv")
```

```python
result_df.head()
```

```python
result_df.label = result_df.label.apply(eval)
```

```python
plt.figure()
plt.title('Distribution of number of assigned topics per tweet')
result_df.label.apply(lambda x: len(x)).hist(bins='auto')
plt.show()
```

```python
def plt_fractions(df):
    fig = plt.figure()
    plt_data = result_df.iloc[:, result_df.columns != "label"].copy()
    plt_data.columns = [x.split(".")[1] for x in plt_data.columns]
    p = plt_data.round().mean().sort_values().plot.barh()
    p.set_xlim(0.0, 1.0)
    plt.title("Fraction of tweets with found topic")
    fig.tight_layout()
    plt.plot();
```

```python
plt_fractions(result_df)
```
